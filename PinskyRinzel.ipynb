```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

```python
#Constants
Gna = 30
Gkdr = 15
Gkca = 15
Gkahp = 0.8
Gca = 10
Gl = 0.1
Gc = 2.1 #1.425
Gnmda = 0
Vna = 120
Vk = -15
Vca = 140
Vl = 0
p = 0.5
Ca = 0.21664282 
Cm = 3
betac = 2 
alphac = 0.1
#tau_AMPA = 2
#gAMPA = 1e-0.6
#VAMPA = 0
Vsyn = 60

Isapp = 0.75
Idapp = 0
```

```python
#Alpha and Beta functions and Chi

def H(X):
    if X < 0:
        return 0
    else: 
        return 1
    
def vtrap(x,y):
        #val=0
        if(np.abs(x/y) < 0.000001):
            val = (1 - (x/(y))/2)
        else:
            val = (x)/(np.exp((x)/y) - 1.0)
        return val

def alphaM(Vs):
#(13.1 - Vs)/(np.exp((13.1 - Vs)/4) - 1)
    #return (0.32 * vtrap(13.1 - Vs,4.00))
    return (0.32 * (13.1 - Vs)/(np.exp((13.1 - Vs)/4) - 1))

def betaM(Vs):
#(Vs - 40.1)/(np.exp((Vs - 40.1)/5) - 1)
    #return (0.28 * vtrap(Vs - 40.1,5.00))
    return (0.28 * (Vs - 40.1)/(np.exp((Vs - 40.1)/5) - 1))

def Minf(Vs):
    return (alphaM(Vs)/(alphaM(Vs) + betaM(Vs)))

def alphaN(Vs):     
#(35.1 - Vs))/(np.exp((35.1 - Vs)/5.00) - 1.00    
    #return ((0.016 * vtrap(35.1 - Vs,5.00)))
    return (0.016 * (35.1 - Vs))/(np.exp((35.1 - Vs)/5.00) - 1.00)

def betaN(Vs):
    return (0.25*np.exp(0.5 - 0.025*Vs))

def alphaH(Vs):
    return (0.128*np.exp((17-Vs)/18))

def betaH(Vs):
    return (4/(1 + np.exp((40-Vs)/5)))

def alphaS(Vd):
    return (1.6/(1 + np.exp(-0.072*(Vd - 65))))

def betaS(Vd):
#(Vd - 51.1)/(np.exp((Vd - 51.1)/5) - 1.00)
    #return (0.02 * vtrap(Vd - 51.1,5.00))
    return 0.02 * (Vd - 51.1)/(np.exp((Vd - 51.1)/5) - 1.00)

#def alphaC(Vd):
#    return ((1 - H(Vd + 10)*np.exp((Vd + 50)/11 - (Vd + 53.5)/27)/18.975) + H(Vd + 10)*(2*np.exp((-53.5-Vd)/27)))

def alphaC(Vd):
    if Vd > 50:
        return (2.00 * np.exp((6.5 - Vd)/27))
    else:
        return ((np.exp((Vd - 10)/11) - np.exp((Vd - 6.5)/27))/18.975)
    
def betaC(Vd):
    if Vd > 50:
        return 0
    else:
        return ((2.00 * np.exp((6.5 - Vd)/27)) - alphaC(Vd))

#return ((1 - H(Vd + 10))*(2*np.exp((-53.5 - Vd)/27) - alphaC(Vd)))

def alphaQ(Ca):
    return (min(0.00002*Ca,0.01))

def betaQ(Ca):
    return (0.001)

def chi(Ca):
    #print (Ca) 
    return min(Ca/250.0,1)
    
def qinf(Ca):
    return ((0.7894 * np.exp(0.0002726*Ca)) - (0.7292 * np.exp(-0.01672*Ca)))

#def tauq(Ca):
#    return (657.9 * np.exp(-0.02023*Ca)) + (301.8 * np.exp(-0.002381*Ca))

def cinf(Vd):
    return ((1.0/(1.0 + np.exp((-10.1 - Vd)/0.1016)))**0.00925)

#def tauc(Vd):
#    return (3.627 * np.exp(0.03704*Vd))

#def chi(Ca):
#    return (1.073*np.sin(0.003453*Ca+0.08095) \
#            + 0.08408*np.sin(0.01634*Ca-2.34)\
#            + 0.01811*np.sin(0.0348*Ca-0.9918))


#RK


def dVdt(V, t):
    #print("t : ",t, "V: ", V)
    Vs, Vd,m, n, h, s, c, q, Ca, Si = V
    
    #Equations for currents
    Ina = Gna * (Minf(Vs))**2 * h * (Vs - Vna) #Mi is minfinity
    Ikdr = Gkdr * n * (Vs - Vk)
    Ica = Gca * s**2 * (Vd - Vca)
    Ikca = Gkca * c * chi(Ca) * (Vd - Vk)
    Ikahp = Gkahp * q * (Vd - Vk)
    Ids = Gc * (Vd - Vs)
    Isd = Gc * (Vs - Vd)
    Ileakd = Gl * (Vd - Vl)
    Ileaks = Gl * (Vs - Vl)
    Inmda = 0#Gnmda*Si*(Vd - Vsyn)/(1 + 0.28*np.exp(-0.062*(Vd - 60)))
    Isyn = 0

    dCadt = (-0.13*Ica - 0.075*Ca)
    dvsdt = (-Ileaks - Ina - Ikdr + Ids/p + Isapp/p)/Cm
    dvddt = (-Ileakd - Ica - Ikca - Ikahp - Isyn/(1-p) + Isd/(1-p) - Inmda/(1-p) + Idapp/(1-p))/Cm
    dmdt = (((alphaM(Vs)/(alphaM(Vs) + betaM(Vs))) - m)/(1/(alphaM(Vs) + betaM(Vs))))
    dndt = (((alphaN(Vs)/(alphaN(Vs) + betaN(Vs))) - n)/(1/(alphaN(Vs) + betaN(Vs))))
    dhdt = (((alphaH(Vs)/(alphaH(Vs) + betaH(Vs))) - h)/(1/(alphaH(Vs) + betaH(Vs))))
    dsdt = (((alphaS(Vd)/(alphaS(Vd) + betaS(Vd))) - s)/(1/(alphaS(Vd) + betaS(Vd))))
    dcdt = (((alphaC(Vd)/(alphaC(Vd) + betaC(Vd))) - c)/(1/(alphaC(Vd) + betaC(Vd))))
    dqdt = (((alphaQ(Ca)/(alphaQ(Ca) + betaQ(Ca))) - q)/(1/(alphaQ(Ca) + betaQ(Ca))))
    #dqdt = (qinf(Ca) - q)/tauq(Ca)
    #dcdt = (cinf(Vd) - c)/tauc(Vd)
    #dqdt = alphaQ(Ca) - (alphaQ(Ca) + betaQ(Ca))*q
    dSidt = H(Vs - 10) - Si/150
    
    
    return dvsdt, dvddt, dmdt, dndt, dhdt, dsdt, dcdt, dqdt, dCadt, dSidt
```

```python
tmin, tmax, dt = 0, 1000, 0.01
T = np.arange(tmin,tmax,dt)

V0 = [-4.6, -4.5, 0.5, 0.001, 0.999, 0.009, 0.007, 0.01, 0.2, 0] #Initital values need to be feeded
sol = odeint(dVdt, V0, T)
```
